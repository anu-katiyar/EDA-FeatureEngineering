```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
train_data = pd.read_csv('Data_Train.csv')
train_data.head()
```

```python
test_data = pd.read_csv('Test_set.csv')
test_data.head()
```

```python
data = pd.concat([train_data,test_data])
```

```python
data.tail()
```

```python
data.isnull().sum()
```

```python
data['Date_of_Journey'].str.split('/').str[1]
```

```python
data['Day'] = data['Date_of_Journey'].str.split('/').str[0]
data['Month'] = data['Date_of_Journey'].str.split('/').str[1]
data['Year'] = data['Date_of_Journey'].str.split('/').str[2]
```

```python
data.head()
```

```python
data['Day']=data['Day'].astype(int)
data['Month']=data['Month'].astype(int)
data['Year']=data['Year'].astype(int)
```

```python
data.info()
```

```python
data.drop('Date_of_Journey', inplace= True, axis=1)
```

```python
data.head()
```

```python
data['Arrival_Time'] = data['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
data.head()
```

```python
data['Arrival_hr'] = data['Arrival_Time'].apply(lambda x: x.split(':')[0])
data['Arrival_Min'] = data['Arrival_Time'].apply(lambda x: x.split(':')[1])
```

```python
data['Arrival_hr'] = data['Arrival_hr'].astype(int)
data['Arrival_Min'] = data['Arrival_Min'].astype(int)
```

```python
data.drop('Arrival_Time', axis =1, inplace=True)
```

```python
data.head()
```

```python
data['Dep_Hr'] = data['Dep_Time'].apply(lambda x: x.split(':')[0])
data['Dep_Min'] = data['Dep_Time'].apply(lambda x: x.split(':')[1])

data['Dep_Hr'] = data['Dep_Hr'].astype(int)
data['Dep_Min'] = data['Dep_Min'].astype(int)
data.drop('Dep_Time', axis =1, inplace=True)
```

```python
data.info()
```

```python
data['Total_Stops'].unique()
```

```python
data['Total_Stops'] = data['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4, 'nan':1})
```

```python
data.drop('Route', axis=1, inplace=True)
```

```python
data.head()
```

```python
data['TotalDuration']=data['Duration'].apply(lambda x: x.split('h ')).str[0]
```

```python
 data['TotalDuration']=data['TotalDuration'].apply(lambda x: x.split('h')).str[0]
```

```python
data.head()
```

```python
# Drop rows with invalid data as these are insignficant rows and won't have impact on analysis and prediction
data.drop(6474, axis = 0, inplace = True)
data.drop(2660, axis = 0, inplace = True)
```

```python
data['TotalDuration'] = data['TotalDuration'].astype(int)
```

```python
data['TotalDuration'] = data['TotalDuration'].apply(lambda x: x*60)
```

```python
data.head()
```

```python
data['TotalDurMin'] = data['Duration'].apply(lambda x: x.split('h ')).str[1].str.split('m').str[0]
```

```python
data['TotalDurMin'].fillna(0, inplace=True)
```

```python
data['TotalDurMin'] = data['TotalDurMin'].astype(int)
```

```python
data.info()
```

```python
data['TotalDuration'] =  data['TotalDuration'] + data['TotalDurMin']
data.drop('Duration', axis =1, inplace=True)
data.drop('TotalDurMin', axis =1, inplace=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Source'] = le.fit_transform( data['Source'])
data['Destination'] = le.fit_transform( data['Destination'])
data['Additional_Info'] = le.fit_transform( data['Additional_Info'])
data['Airline'] = le.fit_transform( data['Airline'])
```
